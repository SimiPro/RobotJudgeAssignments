```python
# load spacy and english dictionary
import spacy

nlp = spacy.load('en')
```

```python
from os import listdir


# load a document 
N = 1500
case_path = 'data/cases'
all_cases = listdir(case_path)
```

```python
NUM_WORDS = []
NUM_SENTECES = []
NUM_LETTERS = []
for i in range(len(all_cases[:N])):
    if (i % 10 == 0): print('{}%'.format(int(100/len(all_cases[:N])*i)), end=', ')
    
    case_0_name = '{}/{}'.format(case_path, all_cases[i])
    case_0 = open(case_0_name, 'r').read()
    doc = nlp(case_0)

    # compute:
        # number of sentences 
    num_sentences = len(list(doc.sents))
    NUM_SENTECES.append(num_sentences)
        # number of words
    words = [token.orth_ for token in doc if not token.is_punct | token.is_space | token.is_digit ]
    num_words = len(words)
    NUM_WORDS.append(num_words)
        # number of letters
    words_cleaned = [w.replace(',', '') for w in words]
    words_cleaned = [len(w.replace('.', '')) for w in words_cleaned]
    num_letters = sum(words_cleaned)
    NUM_LETTERS.append(num_letters)
```

```python
import numpy as np

    
NUM_WORDS = np.array(NUM_WORDS)
NUM_SENTECES = np.array(NUM_SENTECES)
NUM_LETTERS = np.array(NUM_LETTERS)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.title('Histogram over number of Words')
plt.xlabel('Number of Words')
plt.ylabel('Number of Occurences')
plt.hist(NUM_WORDS)
plt.show()

plt.figure()
plt.title('Histogram over number of Sentences')
plt.xlabel('Number of Sentences')
plt.ylabel('Number of Occurences')
plt.hist(NUM_SENTECES)
plt.show()

plt.figure()
plt.title('Histogram over number of Letters')
plt.xlabel('Number of Letters')
plt.ylabel('Number of Occurences')
plt.hist(NUM_LETTERS)
plt.show()
```

```python
# 2. Use the spaCy parts of speech (POS)
# tags to count number of nouns, verbs, and adjectives in each document. 
# Visualize POS frequency by year.
NUM_NOUNS = []
NUM_VERBS = []
NUM_ADJ = []
for i in range(len(all_cases[:N])):
    if (i % 10 == 0): print('{}%'.format(int(100/len(all_cases[:N])*i)), end=', ')

    case_0_name = '{}/{}'.format(case_path, all_cases[i])
    case_0 = open(case_0_name, 'r').read()
    doc = nlp(case_0)
    
    nouns = []
    verbes = []
    adjs = []
    for token in doc:
        if token.pos_ == "NOUN":
            nouns.append(token)
        elif token.pos_ == "VERB":
            verbes.append(token)
        elif token.pos_ == "ADJ":
            adjs.append(token)
    NUM_NOUNS.append(len(nouns))
    NUM_VERBS.append(len(verbes))
    NUM_ADJ.append(len(adjs))
```

```python
# Follow the steps in lecture to normalize your corpus (e.g., removing punctuation) 
# and discuss your choices about what information to exclude. 
# Using the normalized tokens, make a feature set of all trigrams that end in a noun.  



# I already did the exclusion of punctuation etc already for the analysis of 
# letters and words because i thought that words and letters should not include that.

# For this case I just remove [","] i will keep "." and "!" because I try to keep trigrams 
# that stem from the same sentence. 

## Now gather all trigrams that end in a noun
## and count how many times each trigram occurs

def get_trigrams(case):
    doc = nlp(case_0)
    trigrams = []
    current_trigram = []
    for token in doc:
        if token.text == ",": continue
        current_trigram.append(token.text)
        
        if token.pos_ == "NOUN":
            if len(current_trigram) == 3:
                trigrams.append(tuple(current_trigram))
                current_trigram = []
        if len(current_trigram) > 2:
            current_trigram.pop(0)
    return trigrams

all_trigrams = {}
for i in range(len(all_cases[:N])):
    if (i % 10 == 0): print('{}%'.format(int(100/len(all_cases[:N])*i)), end=', ')

    case_0_name = '{}/{}'.format(case_path, all_cases[i])
    case_0 = open(case_0_name, 'r').read()
    
    for ngram in get_trigrams(case_0):
        if ngram not in all_trigrams:
            all_trigrams[ngram]  = 1
        else:
            all_trigrams[ngram] += 1
```

```python
# We take 1000 biggest trigrams as features 
bigger = 0
for (key, val) in all_trigrams.items():
    bigger += (val > 1)

all_trigrams_list = list(all_trigrams.items())

    
print(bigger, " of ", len(all_trigrams))
```

```python
import pandas as pd
case_reversed = pd.read_csv('data/case_reversed.csv')
case_reversed.head()
```

```python
# Generate DATA

import re

i = 0

X = np.zeros((N, 1000))
Y = np.zeros(N)

for i in range(len(all_cases[:N])):
    if (i % 10 == 0): print('{}%'.format(int(100/len(all_cases[:N])*i)), end=', ')

    file_name = all_cases[i]
    case_id = re.findall(r'[A-Z]+[A-Z0-9]*',file_name)[0]
    Y[i] = int(case_reversed.loc[case_reversed['caseid'] == case_id]['case_reversed'])

    case_0_name = '{}/{}'.format(case_path, all_cases[i])
    case_0 = open(case_0_name, 'r').read()
    for ngram in get_trigrams(case_0):
        if ngram in features_map:
            X[i, features_map[ngram]] += 1
```

```python
# SPLIT Data into training and test and run logistic regression on it

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

```python
# Use GridSearchCV to find correct parameters
from sklearn.model_selection import GridSearchCV

parameters = {'penalty':['l2', 'l1'] ,'C':[0.1, 0.5, 1, 5]}
logistic = LogisticRegression(random_state=0, solver='liblinear')

clf = GridSearchCV(logistic, parameters, cv=5)
clf.fit(X_train, y_train)

clf.best_estimator_
```

```python
acc = clf.best_estimator_.score(X_test, y_test)
print("Accuracy: {:.4f}".format(acc))
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

vader_score = pd.DataFrame(columns=['sentence', 'compound', 'neg', 'neu', 'pos'])

sid = SentimentIntensityAnalyzer()

row_counter = 0
for i in range(len(all_cases[:N])):
    if (i % 10 == 0): print('{}%'.format(int(100/len(all_cases[:N])*i)), end=', ')
    file_name = all_cases[i]
    case_0_name = '{}/{}'.format(case_path, all_cases[i])
    case_0 = open(case_0_name, 'r').read()
    doc = nlp(case_0)
    sentences = [str(sent) for sent in list(doc.sents)]   
    for sentence in sentences:
        ss = sid.polarity_scores(sentence)
        vader_score.loc[row_counter] = [sentence, ss['compound'], ss['neg'], ss['neu'], ss['pos']]
        row_counter += 1
```

```python
all_sentences = []
for i in range(len(all_cases[:N])):
    if (i % 10 == 0): print('{}%'.format(int(100/len(all_cases[:N])*i)), end=', ')
    file_name = all_cases[i]
    case_0_name = '{}/{}'.format(case_path, all_cases[i])
    case_0 = open(case_0_name, 'r').read()
    doc = nlp(case_0)
    sentences = [str(sent) for sent in list(doc.sents)]   
    all_sentences.extend(sentences)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import pairwise_distances
#all_case_files = ["{}/{}".format(case_path, case) for case in all_cases[:N]]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(all_sentences)
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import linear_kernel

def find_similar(tfidf_matrix, index, top_n = 5):
    cosine_similarities = linear_kernel(tfidf_matrix[index:index+1], tfidf_matrix).flatten()
    related_docs_indices = [i for i in cosine_similarities.argsort()[::-1] if i != index]
    return [(index, cosine_similarities[index]) for index in related_docs_indices][0:top_n]

print("SOURCE SENTENCES: ")
for s in all_sentences[:10]:
    print(s)

print("SIMILARITIES:")
for index, score in find_similar(tfidf_matrix, 10):
    print(score, all_sentences[index])
```

```python
#### Clustering
from sklearn.cluster import MiniBatchKMeans

N_CLUSTERS = 20
cluster = MiniBatchKMeans(n_clusters=N_CLUSTERS)
cluster.fit(tfidf_matrix)
```

```python
for i in range(N_CLUSTERS):
    print("\n\n SENTENCES OF CLUSTER: {}\n\n".format(i+1))
    j = 0
    for k, lbl in enumerate(cluster.labels_):
        if lbl == i:
            j += 1
            print(all_sentences[k])
        if j >= 5:
            break
        
```
